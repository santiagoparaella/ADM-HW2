```python
#create a list of all premier league teams
premier_league = ["Arsenal",
                  "AFC Bournemouth",
                  "Brighton & Hove Albion",
                  "Burnley","Chelsea",
                  "Crystal Palace",
                  "Everton","Huddersfield Town",
                  "Leicester City",
                  "Liverpool",
                  "Manchester City",
                  "Manchester United",
                  "Newcastle United",
                  "Southampton",
                  "Stoke City","Swansea City",
                  "Tottenham Hotspur",
                  "Watford",
                  "West Bromwich Albion",
                  "West Ham United"]
```

```python
#match the premier league list to teams data
df_teams = df_teams[df_teams['officialName'].str.match('|'.join(premier_league))]
```

```python
df_teams
```

```python
#drop unnecessary columns
df_teams = df_teams.drop(['area', 'name', 'type'], axis=1)
```

```python
df_teams
```

```python
#rename the column names
df_teams.columns = ['city','currentTeamId','officialName']
```

```python
df_teams
```

```python
#read json data into pandas datframe
df_coaches = pd.read_json ('/users/poojacj/Desktop/adm_hw2/coaches.json')
```

```python
df_coaches.head(100)
```

```python
df_coaches.shape
```

```python
df_coaches.columns.values
```

```python
df_coaches.shape
```

```python
#drop unnecessary columns
df_coaches = df_coaches.drop(['birthArea', 'passportArea'], axis=1)
```

```python
df_coaches.head(100)
```

```python
#left join the teams and coches dataframes on id
df_merge_col = pd.merge(df_teams, df_coaches, on='currentTeamId', how='left')
```

```python
df_merge_col.head()
```

```python
#drop the rows with empty date of birth values
df_merge_col = df_merge_col.dropna(axis = 0, how ='any')
```

```python
#convert the birthdate datatype to datetime for calculation of age 
df_merge_col['birthdate_datetime'] = pd.to_datetime(df_merge_col.birthDate, infer_datetime_format=True)
```

```python
df_merge_col['birthdate_datetime']
```

```python
#calculate the age of coaches
df_merge_col['age'] = ((pd.to_datetime('today') - df_merge_col['birthdate_datetime'])/365).dt.days
```

```python
df_merge_col['age']
```

```python
df_merge_col
```

```python
#sort the columns in ascending order of age
df_age_sorted = df_merge_col.sort_values(by='age', ascending=True)
```

```python
#select the top 10 rows
df_youngest_coaches = df_age_sorted[:10]
```

```python
results = df_youngest_coaches[['officialName','shortName','age']]
```

```python
results.columns = ['TeamName','CoachName','CoachAge']
```

```python
results
```

```python
#save the results to a csv file
results.to_csv("/users/poojacj/Desktop/adm_hw2/youngest_coaches.csv",index=None)
```

```python
df_age_sorted
```

```python
plt.figure(figsize=(18,10))
#add boxplot
chart = sns.boxplot(
    data=df_age_sorted,
    x ='officialName',
    y='age',
    palette='Set1'
)
# add swarmplot
chart=sns.swarmplot(
    y='age',
    x='officialName',
    data=df_age_sorted, 
    color='black',
    alpha=0.75,
)
# make changes to x and y labels
chart.set_xticklabels(chart.get_xticklabels(),rotation=30,horizontalalignment='right',
                      fontweight='light',
                      fontsize='x-large')
chart.axes.set_title("Distribution of Age of coaches in Premier League",
                    fontsize=20)
chart.set_xlabel("Teams", 
                fontsize=14)
chart.set_ylabel("Age",
                fontsize=14)
```

```python
# output file name
plot_file_name="/users/poojacj/Desktop/adm_hw2/age_distribution_boxplot.jpg"
# save as jpeg
chart.figure.savefig(plot_file_name,
                    format='jpeg',
                    dpi=100)
```
